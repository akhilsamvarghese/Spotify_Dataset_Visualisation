### **Outlier Detection and Handling**


```python
# Plot boxplots for numerical columns to identify outliers
numerical_columns = df.select_dtypes(include=[np.number]).columns
fig, axes = plt.subplots(nrows=len(numerical_columns), figsize=(10, 5*len(numerical_columns)))
for i, col in enumerate(numerical_columns):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col)
plt.tight_layout()
plt.show()
```

```python
# After identifying outliers, the next steps are to analyze their impact and decide how to handle them.
# We'll use the Interquartile Range (IQR) method to detect outliers and then create a function to handle them.

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

def handle_outliers(df, column, method='clip'):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    if method == 'clip':
        return df[column].clip(lower_bound, upper_bound)
    elif method == 'remove':
        return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    else:
        raise ValueError("Method must be either 'clip' or 'remove'")

# Analyze outliers for each numerical column
for column in numerical_columns:
    outliers = detect_outliers(df, column)
    print(f"Outliers in {column}: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
    
    # Handle outliers (here we're using the 'clip' method, but you can change it to 'remove' if needed)
    df[column] = handle_outliers(df, column, method='clip')

# Verify that outliers have been handled
fig, axes = plt.subplots(nrows=len(numerical_columns), figsize=(10, 5*len(numerical_columns)))
for i, col in enumerate(numerical_columns):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f"{col} (after handling outliers)")
plt.tight_layout()
plt.show()

# Print summary statistics after handling outliers
print(df[numerical_columns].describe())
```
