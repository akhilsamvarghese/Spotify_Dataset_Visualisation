# **Pre-processing the Dataset** 
*****

### **Check for missing values**

```python

print(df.isnull().sum())
```

```python
# Drop the 'in_shazam_charts' column
df = df.drop(columns=['in_shazam_charts'])
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["key"])
```

```python
#Checking if the missing values are dropped
print(df.isnull().sum())
```

```python
# Check for duplicates
print(df.duplicated().sum())
```

# **Convert data types:**

```python

# Convert 'streams' column to integer type
# Remove rows where 'streams' column contains non-numeric values
df = df[pd.to_numeric(df['streams'], errors='coerce').notnull()]

# Convert 'streams' column to integer type
df['streams'] = df['streams'].astype(int)
```

```python
df.info()
```

```python
# Renaming the columns 
df = df.rename(columns={
    'artist(s)_name': 'artists',
    'released_year': 'year',
    'released_month': 'month',
    'released_day': 'day'
})
```

```python
df.head()
```

```python
# Create release_date column
df['release_date'] = pd.to_datetime(df[['year', 'month', 'day']])
```

```python
# Remove unnecessary columns
columns_to_drop = ['year', 'month', 'day']  # Add any other columns you don't need
df = df.drop(columns=columns_to_drop)

```

```python
df.info()
```

### **Exporting the cleaned data-frame into a csv file** 


```python
df.to_csv("cleaned_spotify_data.csv", index=False)
```

```python
# 
plt.figure(figsize=(18, 6))
sns.scatterplot(x='danceability_%', y='energy_%', data=df, alpha=0.6)
plt.title('Danceability vs Energy')
plt.xlabel('Danceability (%)')
plt.ylabel('Energy (%)')
plt.show()
```

```python
#Violin plot of Valence by in_spotify_charts
plt.figure(figsize=(28, 6))
sns.violinplot(x='in_spotify_charts', y='valence_%', data=df)
plt.title('Distribution of Valence for Songs In/Out of Spotify Charts')
plt.xlabel('In Spotify Charts')
plt.ylabel('Valence (%)')
plt.show()
```

```python
# **2. Top 10 Artists by Stream Count**
top_artists = df.groupby('artists')['streams'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(16, 8))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='husl')
plt.title('Top 10 Artists by Stream Count')
plt.xlabel('Total Streams')
plt.ylabel('Artists')
plt.show()

```

```python
Top_songs = df[['track_name','streams']]
Top_songs['streams']=pd.to_numeric(Top_songs['streams'], errors= 'coerce')
Top= Top_songs[['track_name','streams']].sort_values(by = 'streams', ascending= False).head(10)
Top
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x =Top['streams'],y= Top['track_name'], palette= 'husl')
plt.xlabel('Streams')
plt.ylabel('Songs')
plt.title('Most Streamed Songs on Spotify')
plt.show()
```

```python
### **Top 10 Streamed Songs in 2023**
plt.figure(figsize = (20, 10))

sns.barplot(x = df.track_name[:10], y = df.streams)

plt.title('Top 10 Streamed Songs in 2023')
plt.xlabel('Artist')
plt.ylabel('Streams')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# **3. Top 10 Artists by Number of Songs in the Dataset**
top_artists_by_songs = df['artists'].value_counts().head(10)
plt.figure(figsize=(16, 8))
sns.barplot(x=top_artists_by_songs.values, y=top_artists_by_songs.index, palette='husl')
plt.title('Top 10 Artists by Number of Songs in the Dataset')
plt.xlabel('Number of Songs')
plt.ylabel('Artists')
plt.show()
```

```python
#the barplot represents Relationship Between The Number Of Atrists In A Song and Streams In 2023
plt.figure(figsize = (20, 10))

sns.barplot(x = df.artist_count, y = df.streams)

plt.title('Relationship Between The Number Of Atrists In A Song and Streams In 2023')
plt.xlabel('Number of Artists')
plt.ylabel('Streams')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#Line plot of Average Streams by Release Month
df['release_month'] = df['release_date'].dt.month
monthly_streams = df.groupby('release_month')['streams'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='release_month', y='streams', data=monthly_streams, marker='o')
plt.title('Average Streams by Release Month')
plt.xlabel('Month')
plt.ylabel('Average Streams')
plt.xticks(range(1, 13))
plt.show()
```

```python
# Correlation Heatmap of Audio Features
plt.figure(figsize=(19,15 ))
correlation_matrix = df[['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap of Audio Features')
plt.show()
```
